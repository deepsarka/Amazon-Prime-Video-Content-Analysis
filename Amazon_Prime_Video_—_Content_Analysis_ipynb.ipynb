## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Drop columns that are not useful for EDA
# imdb_id   → external ID, no analytical value
# description → text data, not used in this EDA
titles.drop(columns=['imdb_id', 'description'], inplace=True)
````

````python
# Handle missing values
# - imdb_score  → fill with median (avoids skew from outliers)
# - imdb_votes  → fill with 0 (no votes recorded)
# - tmdb_score  → fill with median
# - tmdb_popularity → fill with median
# - seasons     → fill with 0 (means it's a MOVIE, not a SHOW)
# - age_certification → fill with 'Not Rated'
# - character (credits) → fill with 'Unknown'
titles['imdb_score'].fillna(titles['imdb_score'].median(), inplace=True)
titles['imdb_votes'].fillna(0, inplace=True)
titles['tmdb_score'].fillna(titles['tmdb_score'].median(), inplace=True)
titles['tmdb_popularity'].fillna(titles['tmdb_popularity'].median(), inplace=True)
titles['seasons'].fillna(0, inplace=True)
titles['age_certification'].fillna('Not Rated', inplace=True)
credits['character'].fillna('Unknown', inplace=True)
````

````python
# Merge both the data set
new_data = pd.merge(titles, credits, on='id', how='left')
````

````python
# Era classification (5 time bands)
def get_era(year):
    if year < 1980:   return 'Classic (pre-1980)'
    elif year < 2000: return '80s-90s'
    elif year < 2010: return '2000s'
    elif year < 2016: return '2010-2015'
    else:             return '2016-2022'

titles['era'] = titles['release_year'].apply(get_era)
print('Era distribution:')
print(titles['era'].value_counts())
````

#### Chart - 1

````python
# CHART 1 — UNIVARIATE | Count Plot + Pie Chart
# Variable: type (MOVIE vs SHOW)
# Count values
type_counts = titles['type'].value_counts()

# Create figure with 2 charts
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# LEFT: Count Plot
sns.countplot(data=titles, x='type', ax=axes[0], palette='husl')

axes[0].set_title("Content Type Count")
axes[0].set_xlabel("Type")
axes[0].set_ylabel("Number of Titles")

# Add values on top of bars
for bar in axes[0].patches:
    axes[0].text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        int(bar.get_height()),
        ha='center',
        va='bottom'
    )

# RIGHT: Pie Chart
axes[1].pie(
    type_counts.values,
    labels=type_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

axes[1].set_title("Content Type Percentage")

# Layout fix
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A **Count Plot** was chosen because it directly reads raw categorical data and automatically counts how many times each category appears — no pre-calculation needed. This makes it the most appropriate and correct chart for understanding the distribution of a categorical variable like content type. The **Pie Chart** alongside it shows the same data as a proportion, giving a complete picture of the content type balance on the platform.

#### Chart - 4

````python
# Chart - 4 visualization code
# director name × average IMDb score
# Filter directors who have worked on at least 3 titles (more reliable average)
directors_df = new_data[new_data['role'] == 'DIRECTOR'].copy()

director_stats = (
    directors_df.groupby('name')
    .agg(avg_score=('imdb_score', 'mean'), title_count=('title', 'nunique'))
    .reset_index()
)
# Keep directors with at least 3 titles for reliability
top_directors = (
    director_stats[director_stats['title_count'] >= 3]
    .sort_values('avg_score', ascending=False)
    .head(15)
)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_directors['name'], top_directors['avg_score'],
                color=sns.color_palette('Blues_r', len(top_directors)))
for bar, cnt in zip(bars, top_directors['title_count']):
    plt.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.2f} ({cnt} titles)', va='center', fontsize=8)

plt.title('Top 15 Directors by Average IMDb Score (min. 3 titles)', fontsize=13)
plt.xlabel('Average IMDb Score')
plt.ylabel('Director')
plt.xlim(0, 10.5)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A **horizontal bar chart** ranks directors by average IMDb score. Filtering for directors with at least 3 titles ensures the averages are meaningful rather than based on a single lucky title.

##### 2. What is/are the insight(s) found from the chart?

A clear set of directors consistently delivers above-average content quality. The top directors average 7.5–8.5 IMDb scores across multiple titles, showing that talent quality is repeatable and not random. This is a strong signal that **director track record is a reliable quality predictor**.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

**Positive:** Knowing which directors consistently produce high-quality content allows Amazon to prioritize partnerships, acquisition deals, and original production investment with proven talent.

**Negative:** Over-relying on a small group of directors creates risk — if those relationships end, content quality could dip significantly.

#### Chart - 5

````python
# Chart - 5 visualization code
# actor name × average IMDb score (min. 5 appearances)
actors_df2 = new_data[new_data['role'] == 'ACTOR'].copy()

actor_stats = (
    actors_df2.groupby('name')
    .agg(avg_score=('imdb_score', 'mean'), appearances=('title', 'nunique'))
    .reset_index()
)
top_actors = (
    actor_stats[actor_stats['appearances'] >= 5]
    .sort_values('avg_score', ascending=False)
    .head(15)
)

plt.figure(figsize=(10, 6))
bars = plt.barh(top_actors['name'], top_actors['avg_score'],
                color=sns.color_palette('Oranges_r', len(top_actors)))
for bar, cnt in zip(bars, top_actors['appearances']):
    plt.text(bar.get_width() + 0.02, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.2f} ({cnt} titles)', va='center', fontsize=8)

plt.title('Top 15 Actors by Average IMDb Score (min. 5 appearances)', fontsize=13)
plt.xlabel('Average IMDb Score')
plt.ylabel('Actor')
plt.xlim(0, 10.5)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A **horizontal bar chart** with color-coded quality tiers (green = above 6.5, orange = 6.0–6.5, red = below 6.0) gives an instant quality ranking of all genres. Filtering for genres with at least 50 titles ensures statistical reliability.

##### 2. What is/are the insight(s) found from the chart?

**History and Documentary genres average the highest IMDb scores** (typically above 7.0). **Horror and Reality content average the lowest**. Drama and Thriller sit in the mid-tier. This reveals a clear quality hierarchy across content categories that is invisible from volume charts alone.

##### 3. Will the gained insights help creating a positive business impact?

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# all numeric columns together
# Select numeric columns
numeric_cols = ['release_year', 'runtime', 'seasons',
                'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']

# Correlation matrix
corr = titles[numeric_cols].corr()

# Mask upper triangle
mask = np.triu(np.ones_like(corr), k=1)

# Plot heatmap
plt.figure(figsize=(10,7))

sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    cmap='coolwarm',
    center=0
)

plt.title("Correlation Heatmap")
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is the most efficient way to display pairwise relationships between multiple numerical variables at once. Instead of drawing dozens of individual scatter plots, a single heatmap encodes every correlation coefficient as a color intensity — red for positive, blue for negative — letting analysts instantly spot which variable pairs are strongly or weakly related. It's the standard first-step diagnostic tool in any multivariate data exploration, especially when deciding which variables are worth investigating further.
